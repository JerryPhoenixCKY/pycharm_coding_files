## äº”ã€å®é™…åº”ç”¨ç¤ºä¾‹

### ç¤ºä¾‹ 1ï¼šå€’è®¡æ—¶åˆ°åœ£è¯èŠ‚ï¼ˆç”¨ `datetime`ï¼‰

```python
from datetime import datetime, date

today = date.today()
christmas = date(today.year, 12, 25)
if today > christmas:
    christmas = date(today.year + 1, 12, 25)

diff = christmas - today
print(f"è·ç¦»åœ£è¯èŠ‚è¿˜æœ‰ {diff.days} å¤©ï¼")
```

### ç¤ºä¾‹ 2ï¼šå‡½æ•°æ‰§è¡Œè®¡æ—¶ï¼ˆç”¨ `time`ï¼‰

```python
import time

def slow_function():
    time.sleep(1.2)

start = time.perf_counter()
slow_function()
elapsed = time.perf_counter() - start
print(f"æ‰§è¡Œè€—æ—¶: {elapsed:.2f} ç§’")
```
