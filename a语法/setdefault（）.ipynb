### åœºæ™¯ï¼šæŠŠå•è¯æŒ‰é¦–å­—æ¯åˆ†ç»„

```python
words = ['apple', 'ant', 'banana', 'bear', 'cat']
groups = {}

for word in words:
    first_letter = word[0]
    groups.setdefault(first_letter, []).append(word)

print(groups)
# è¾“å‡º: {'a': ['apple', 'ant'], 'b': ['banana', 'bear'], 'c': ['cat']}
```
